```python
# addning new column
df.withColumn("GrandExp", df["age"] * df["experience"]).show()
```

At that moment, Spark executes the logical plan. It reads the original 1 TB dataset, applies the `withColumn()` transformation to calculate the new `GrandExp` column, and produces the required result. However, it does **not** automatically store this new dataset anywhere. The computed data exists only for that action unless you explicitly cache it or write it to storage.

If the action is:

* `df.show()` → Spark computes the data and displays only the requested rows.
* `df.count()` → Spark computes the data only to count the rows.
* `df.collect()` → Spark computes the data and sends the results to the driver.
* `df.write.parquet("output/")` → Spark computes the data and writes the complete dataset, including the new column, to disk.

Unless you use `cache()`, `persist()`, or `write()`, the computed data is discarded after the action finishes. If another action is executed later, Spark will read the original data again and recompute the transformation.
